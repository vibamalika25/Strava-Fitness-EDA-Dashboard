# **Project Summary -** FitTrack: Strava Fitness Analysis Dashboard



The Strava Fitness Analysis project conducts a comprehensive exploratory data analysis of user fitness metrics from the Fitabase dataset, encompassing daily activity, calorie expenditure, intensity levels, and sleep patterns. The primary goal is to identify behavioral trends, uncover correlations between physical activity metrics, and derive actionable insights that can inform business decisions for fitness platforms. The dataset—comprising merged CSV files covering a 30-day period—contains over 30 variables, including total steps, distance traveled (tracked, logged, and active), intensity-specific minutes (very active, fairly active, lightly active, and sedentary), and complementary data like sleep records and weight logs. Given the richness of the data, the analysis systematically explores univariate, bivariate, and multivariate dimensions to paint a complete picture of user fitness behavior.

The univariate analysis reveals typical patterns in step counts and activity distances, with most users clustering in moderate activity ranges while a few exhibit extremely high step or distance counts, indicating distinct user segments, from casual walkers to highly active individuals. Bivariate explorations then demonstrate that step counts and total distance are strongly correlated, as expected, but more nuanced findings emerge when mapping intensity-specific activity against calories burned: very active minutes show the strongest positive correlation with caloric expenditure, whereas sedentary minutes exhibit an inverse relationship, confirming that higher inactivity levels meaningfully reduce daily energy burn. The relationship between tracked and logged activity distances is also noteworthy, as it shows that users frequently under-report logged activities compared to actual tracked distances, a gap that could inform app-level nudges to improve logging behavior. Additionally, time-series plots across the month reveal weekday–weekend patterns, with step counts and distances dipping on Sundays and peaking midweek—information that can guide the timing of motivational notifications or challenge launches.

The multivariate analysis deepens the understanding of these interactions through correlation heatmaps and pair plots that assess twelve key numeric variables, including activity minutes, distances, and step metrics. The heatmap confirms that distance-based and step-based metrics cluster tightly together, while the two sedentary minute variables (from the main dataset and an auxiliary source) show perfect correlation, validating their interchangeability. Crucially, the pair plots reveal that very active distance and fairly active minutes both independently predict total daily calorie burn, but their combined effect is larger than either alone—a finding that can be used to promote balanced activity routines rather than focusing on a single intensity level. The analysis also identifies multicollinearity among variables such as TotalDistance, TrackerDistance, and StepTotal, advising dimensionality reduction or feature selection in any future predictive modeling.

The resulting visualizations—including interactive treemaps, correlation heatmaps, and dynamic bar charts—provide both immediate and longitudinal insights for business stakeholders. For a fitness platform, these insights can drive product improvements by enabling personalized goal-setting based on user activity profiles; for example, users with high sedentary minutes can be offered gentle reminders to increase light activity, while those who already log moderate to high intensity can be encouraged to sustain or diversify their routines. Moreover, the identified weekend–weekday differences support the design of weekly engagement campaigns, such as "weekend warrior" challenges that capitalize on natural activity patterns or weekday step-streak competitions to maintain momentum. The strong correlations between distance, steps, and calories also allow for more accurate energy expenditure predictions even when only step count data is available—a practical consideration for platforms with limited sensor data.

Overall, this project not only fulfills its analytical objectives but also translates data patterns into strategic business opportunities. It underscores the importance of understanding user activity segments, intensity compositions, and temporal habits to foster retention and deepen engagement. Future work could extend this analysis with machine learning models to predict churn based on declining activity metrics or to recommend personalized workout plans. Ultimately, the Strava Fitness Analysis Dashboard serves as a foundational tool for transforming raw activity logs into meaningful narratives that enhance user experience and drive positive health outcomes.

### Dataset First View

````python
import pandas as pd
from functools import reduce

datasets=[dataset1,dataset2,dataset3,dataset4,dataset5,dataset6]
date_cols_options = ['ActivityDate', 'ActivityDay', 'SleepDay', 'Date']
common_date_col_name = 'ActivityDate'

def standardize_and_process_df(df_input):
    df = df_input.copy()
    found_date_col = None
    for col in date_cols_options:
        if col in df.columns:
            found_date_col = col
            break
    if found_date_col:
        if found_date_col != common_date_col_name:
            df = df.rename(columns={found_date_col: common_date_col_name})
        df[common_date_col_name] = pd.to_datetime(df[common_date_col_name], errors='coerce')
        df.dropna(subset=[common_date_col_name], inplace=True)

    return df
processed_datasets = [standardize_and_process_df(ds) for ds in datasets]
md = processed_datasets[0]
for i in range(1, len(processed_datasets)):
    df_to_merge = processed_datasets[i]
    current_merge_keys = ['Id']
    if common_date_col_name in md.columns and common_date_col_name in df_to_merge.columns:
        current_merge_keys.append(common_date_col_name)
    md = pd.merge(md, df_to_merge, on=current_merge_keys, how='outer', suffixes=('', f'_from_ds{i+1}'))
````

### Variables Description


The dataset contains **34 variables** spanning user identification, activity tracking, intensity metrics, and supplementary health data. Below is a comprehensive description of each variable:

## **User Identification**
- **Id**: Unique numeric identifier for each user (33 distinct users). Serves as the primary key for user-level aggregation.

## **Temporal Variable**
- **ActivityDate**: Date of recorded activity (format: YYYY-MM-DD). All activities are recorded from April–May 2016, with one record per user per day.

## **Step Metrics**
- **TotalSteps**: Total number of steps taken in a day. This is the primary indicator of daily physical activity. Range: 0–23,186 steps.
- **StepTotal**: Redundant variable mirroring `TotalSteps` (likely from an auxiliary dataset). Shows identical values.

## **Distance Metrics**
- **TotalDistance**: Total distance covered in a day (in kilometers/miles, unit depends on user settings). Range: 0–20.40 units.
- **TrackerDistance**: Distance recorded by the fitness tracker. Nearly identical to `TotalDistance` (correlation ~0.99), confirming tracker accuracy.
- **LoggedActivitiesDistance**: Distance logged manually or via external applications. Often zero, indicating that most users rely on automatic tracking.

## **Intensity-Specific Distance (Primary Dataset)**
- **VeryActiveDistance**: Distance covered during high-intensity activity (beyond 6 METs equivalent).
- **ModeratelyActiveDistance**: Distance covered during moderate-intensity activity (3–6 METs).
- **LightActiveDistance**: Distance covered during light-intensity activity (1.5–3 METs).
- **SedentaryActiveDistance**: Distance covered during sedentary activity (rare; mostly zero).

## **Intensity-Specific Minutes (Primary Dataset)**
- **VeryActiveMinutes**: Minutes of high-intensity physical activity.
- **FairlyActiveMinutes**: Minutes of moderate-intensity physical activity.
- **LightlyActiveMinutes**: Minutes of light-intensity physical activity.
- **SedentaryMinutes**: Minutes spent in sedentary behavior (average ~950 minutes/day, confirming high inactivity).

## **Intensity Metrics (Auxiliary Dataset)**
- **Variables with `_from_ds3` suffix**: Duplicates of the above intensity minutes and distance variables (e.g., `SedentaryMinutes_from_ds3`, `LightActiveDistance_from_ds3`). These are included for completeness but are functionally redundant.

## **Calorie & Energy Metrics**
- **Calories**: Estimated daily caloric expenditure (in kcal). Range: 0–4900 kcal. Derived from activity intensity and user profile.
- **Calories_from_ds2**: Calorie data from a secondary dataset. Highly correlated with the primary `Calories` variable.

## **Sleep Metrics**
- **TotalSleepRecords**: Number of sleep sessions recorded (typically 1–3).
- **TotalMinutesAsleep**: Total minutes of actual sleep per day.
- **TotalTimeInBed**: Total minutes spent in bed (including awake time). Slightly higher than `TotalMinutesAsleep` due to sleep latency/wakefulness.

## **Body Composition Metrics**
- **WeightKg**: User weight in kilograms (sparse: only 67 non-null records).
- **WeightPounds**: User weight in pounds (derived from WeightKg; 2.20462 conversion factor).
- **Fat**: Body fat percentage (extremely sparse: only 2 non-null records).
- **BMI**: Body Mass Index (sparse: 67 non-null records).

## **Meta Variables**
- **IsManualReport**: Boolean flag indicating whether weight data was manually entered (67 records).
- **LogId**: Timestamp for the log entry (67 records).
